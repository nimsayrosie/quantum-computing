```python
import qiskit
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from numpy import pi
from math import sqrt
import qiskit.quantum_info as qi
from qiskit.visualization import plot_histogram
from qiskit_aer import AerSimulator
from qiskit import transpile
```

# Circuito de 1 Qubit

```python
quantum_circuito1 = QuantumCircuit(1) # Criar um circuito quântico com 1 qubit.
quantum_circuito1.h(0) # Aplicar a porta Hadamard ao qubit, colocando-o em uma superposição de |0⟩ e |1⟩.
quantum_circuito1.measure_all() # Medir o estado do qubit, colapsando a superposição para um dos estados base (|0⟩ ou |1⟩).
quantum_circuito1.draw('mpl') # Visualizar o circuito quântico usando Matplotlib.
```

```python
simulador = AerSimulator() # Criar um simulador quântico usando o Aer.
compilado_circuito = transpile(quantum_circuito1, simulador) # Compilar o circuito quântico para o simulador.
resultado = simulador.run(compilado_circuito).result() # Executar o circuito quântico no simulador e obter os resultados.
```

```python
cont = resultado.get_counts() # Obter a contagem de resultados para o circuito quântico.
plot_histogram(cont) # Visualizar os resultados usando um histograma.
```

# Circuito de 2 Qubit

```python
quantum_circuito2 = QuantumCircuit(2) 
quantum_circuito2.h(0) 
quantum_circuito2.cx(0, 1) 
quantum_circuito2.measure_all() 
quantum_circuito2.draw('mpl')
```

```python
# Criar um simulador quântico usando o Aer, compilar o circuito quântico para o simulador, executar o circuito quântico no simulador e obter os resultados.
simulador2 = AerSimulator() 
compilado_circuito2 = transpile(quantum_circuito2, simulador2) 

# Executar o circuito quântico no simulador e obter os resultados.
cont2 = resultado2.get_counts() 
plot_histogram(cont2) 
```
