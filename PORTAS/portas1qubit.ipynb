# Aplicando porta Pauli-X 

Pauli-X rota 180 graus no eixo X da esfera de bloch, ou seja se o estado for |0> ele inverte para |1>

```python
circuito = QuantumCircuit(1) # Cria um circuito quântico com 1 qubit |0> 
circuito.x(0) # Aplica a porta X no qubit 0
circuito.draw('mpl', scale=2)
```

```python
psi = qi.Statevector(circuito)
print(psi)
```

```python
prob = psi.probabilities()
print(prob)
```

```python
cont = psi.sample_counts(shots=1024)
print(cont)
```

```python
plot_histogram(cont, figsize=(5, 5))
```

```python
psi.draw('bloch', figsize=(5, 5))
```

# Porta Pauli-Y

Rota 180 graus no eixo Y. combina a inversão de estado com uma alteração na fase quântica, transformando ∣0⟩ em 𝑖∣1⟩ e ∣1⟩ em −𝑖∣0⟩ (onde 𝑖 é a unidade imaginária).

```python
circuito2 = circuito # Cria um circuito quântico com 1 qubit |0>
circuito2.y(0) # Aplica a porta Y no qubit 0
circuito2.draw('mpl', scale=2)
```

```python
psi2 = qi.Statevector(circuito2)
print(psi2)
```

```python
prob2 = psi2.probabilities()
print(prob2)
```

```python
cont2 = psi2.sample_counts(shots=1024)
print(cont2)
```

```python
plot_histogram(cont2, figsize=(5, 5))
```

```python
psi2.draw('bloch', figsize=(5, 5))
```

# Porta Pauli-Z

Rota 180 graus no eixo Z.

Ela altera a fase do estado quântico, deixando ∣0⟩ inalterado e transformando ∣1⟩ em −∣1⟩.

```python
circuito3 = QuantumCircuit(1) # Cria um circuito quântico com 1 qubit |0>
circuito3.z(0) # Aplica a porta Z no qubit 0
circuito3.draw('mpl', scale=2)
```

```python
psi3 = qi.Statevector(circuito3)
print(psi3)
```

```python
prob3 = psi3.probabilities()
print(prob3)
```

```python
plot_histogram(cont2, figsize=(5, 5))
```

```python
psi3.draw('bloch', figsize=(5, 5)) # A porta Z não altera a posição do qubit na esfera de Bloch, apenas muda a fase relativa entre os estados |0> e |1>. Portanto, o qubit permanece no estado |0>, que está localizado no polo norte da esfera de Bloch.
```

# Porta Hadamard
 Aplicando a porta em estado de superposição uniforme (0.5, 0.5)
 Aplicando a porta em um estado sem superposição

```python
# Criando um circuito quântico com um qubit
qubit = QuantumRegister(1, 'q')
circuito5 = QuantumCircuit(qubit)

alpha = 1/sqrt(2)  
beta = 1/sqrt(2) 

psi = [alpha, beta]
circuito5.initialize(psi, qubit[0])

circuito5.draw('mpl', scale=2) 
```

```python
psi5 = qi.Statevector(circuito5)
print(psi5)
```

```python
psi5.draw('bloch', figsize=(5, 5))
```

```python
circuito_h = circuito5.copy() # Cria uma cópia do circuito5 para aplicar a porta H
circuito_h.h(0) # Aplica a porta H no qubit 0
circuito_h.draw('mpl', scale=2)
```

```python
psi_h = qi.Statevector(circuito_h)
print(psi_h)
```

```python
psi_h.draw('bloch', figsize=(5, 5)) #Aplicando a porta H no estado de superposição uniforme resulta no estado |0>, que está localizado no polo norte da esfera de Bloch.
```

Portanto, aplicar uma segunda porta Hadamard em um qubit que já está em uma superposição uniforme resultado no estado base ∣ 0 ⟩.

H ( 1/√2   ( |0⟩ + |1⟩)) --> |0⟩  

A porta Hadamard é muitas vezes chamada de "transformada de Fourier discreta de um bit". Quando aplicada a um qubit que já está em superposição, ela atua como uma transformação inversa.

```python
qubit = QuantumRegister(1, 'q') # Cria um circuito quântico com 1 qubit |0>
circuito6 = QuantumCircuit(qubit)

circuito6.draw('mpl', scale=2)
```

```python
psi6 = qi.Statevector(circuito6)    
print(psi6)
```

```python
psi6.draw('bloch', figsize=(5, 5)) # O estado |0> está localizado no polo norte da esfera de Bloch.
```

```python
aplicando_h = circuito6.copy()
aplicando_h.h(0)

aplicando_h.draw('mpl', scale=2)
```

```python
psi7 = qi.Statevector(aplicando_h)
print(psi7)
```

```python
psi7.draw('bloch', figsize=(5, 5)) # O estado está em superposição uniforme, localizado no equador da esfera de Bloch.
```
